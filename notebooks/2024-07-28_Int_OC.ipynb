```r
# import all datasets
data_1 <- readRDS(file = "/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/GSE184880_myeloid_int.RDS")
data_2 <- readRDS(file = "/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/GSE213243_myeloid_int.RDS")
data_3 <- readRDS(file = "/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/GSE217517_myeloid_int.RDS")
data_4 <- readRDS(file = "/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/GSE180661_myeloid_int.RDS")
data_5 <- readRDS(file = "/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/PRJCA005422_myeloid_int.RDS")
```

```r
#merge all datasets
Data <- merge(data_1, y = c(data_2, data_3, data_4, data_5), project = "integration_OC")
```

```r
#Run pca
Data <- RunPCA(Data)
```

```r
#Find neighbours 
Data <- FindNeighbors(Data, dims = 1:30, reduction = "pca")
```

```r
#Cluster
Data <- FindClusters(Data, resolution = 2, cluster.name = "unintegrated_clusters")
```

```r
#Run UMAP
Data <- RunUMAP(Data, dims = 1:30, reduction = "pca", reduction.name = "umap.unintegrated", verbose = FALSE)
```
