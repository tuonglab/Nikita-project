```python
import os
import re
import scanpy as sc

from pathlib import Path
from tqdm import tqdm

sc.logging.print_header()
```

```python
folder = Path("/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/")
files = [f for f in folder.glob("*.h5ad") if f.stem.count("_") == 2]
adatas1 = []
for f in tqdm(files):
    tmp = sc.read_h5ad(f)
    # remove the genes that did not format properly
    keep = [x for x in tmp.var.index if not re.search("[0-9].*.Mar.*|[0-9].*.Sep.*|[0-9].*.Dec.*", x)]
    tmp = tmp[:, tmp.var.index.isin(keep)].copy()
    tmp.layers["counts"] = tmp.X.copy()
    adatas1.append(tmp)
```

```python
import pandas as pd

meta = pd.read_csv(folder / "GSE180661_myeloid_int_metadata.csv")
meta.reset_index(inplace = True)
meta.set_index("level_0", inplace = True)
meta
```

```python
folder = Path("/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/")
files = [f for f in folder.glob("*.h5ad") if f.stem.count("_") > 2]
adatas2 = []
for f in tqdm(files):
    tmp = sc.read_h5ad(f)
    # remove the genes that did not format properly
    keep = [x for x in tmp.var.index if not re.search("[0-9].*.Mar.*|[0-9].*.Sep.*|[0-9].*.Dec.*", x)]
    tmp = tmp[:, tmp.var.index.isin(keep)].copy()
    tmp.layers["counts"] = tmp.X.copy()
    for col in meta:
        tmp.obs[col] = pd.Series(meta[col])
    adatas2.append(tmp)
```

```python
# first concatenate the 200k object
adatax = sc.concat(adatas2)
adatax
```

```python
# concatenate with only intersecting genes
adata = sc.concat(adatas1 + [adatax])
adata
```

```python
# drop seurat_clusters column
adata.obs = adata.obs.drop("seurat_clusters", axis=1)
```

```python
adata.write_h5ad("/scratch/user/s4436039/scdata/Myeloid_Cells/Myeloid_Cells_Integrate/all_integrated_intersect.h5ad", compression = "gzip")
```
