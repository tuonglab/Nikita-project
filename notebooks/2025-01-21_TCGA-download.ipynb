````r
getProjectSummary('TCGA-STAD')
````

````r
#build a query (GDCquery)
query_TCGA <- GDCquery(project = 'TCGA-STAD', 
                       data.category = 'Transcriptome Profiling',
                       experimental.strategy = 'RNA-Seq',
                       workflow.type = 'STAR - Counts',
                       access = 'open')
````

````r
#download data (GDCdownload)
GDCdownload(query_TCGA)
````

````r
#prepare data (GDCprepare)
tcga_stad_data <- GDCprepare(query_TCGA, summarizedExperiment = TRUE)
````

````r
#export randomised summarized experiment
saveRDS(tcga_stad_data, file = "TCGA_STAD.rds")
````
