## normalising the pseudobulk tables so they can be compared when do correlation between DC in different cancers

```python
#load packages
import scanpy as sc
import os
import pandas as pd
import anndata as ad
```

```python
#change current directory 
os.chdir('/scratch/user/s4436039/scdata/Pseudobulk')
os.getcwd()
```

```python
# read in data
pseudobulk_DC1_df = pd.read_csv("pseudobulk_DC1_data.csv", index_col=0)
pseudobulk_DC2_df = pd.read_csv("pseudobulk_DC2_data.csv", index_col=0)
pseudobulk_mregDC_df = pd.read_csv("pseudobulk_mregDC_data.csv", index_col=0)
```

```python
pseudobulk_DC2_df.head()
```

### Normalise:

```python
#normalise to 1e6 --> tested it and this method gives the same as sc.pp.normalize_total() but doesn't require converting to Anndata and back
norm_DC1 = pseudobulk_DC1_df.div(pseudobulk_DC1_df.sum(axis=0), axis=1) * 1e6
norm_DC2 = pseudobulk_DC2_df.div(pseudobulk_DC2_df.sum(axis=0), axis=1) * 1e6
norm_mregDC = pseudobulk_mregDC_df.div(pseudobulk_mregDC_df.sum(axis=0), axis=1) * 1e6
```

```python
norm_DC2.head()
```

### export CSV:

```python
norm_DC1.to_csv("pseudobulk_DC1_norm.csv", index=True)
norm_DC2.to_csv("pseudobulk_DC2_norm.csv", index=True)
norm_mregDC.to_csv("pseudobulk_mregDC_norm.csv", index=True)
```
