```python
import numpy as np
```

```python
# Inspect the data range
print("Range of .X values:", np.min(data.X), "to", np.max(data.X))
```

```python
# Extract the dataset_id column
dataset_ids = data.obs['dataset_id']
```

```python
X_dense = data.X
```

```python
# Loop through each unique dataset_id and calculate the range
for dataset_id in dataset_ids.unique():
    # Get the indices of cells belonging to the current dataset_id
    indices = dataset_ids == dataset_id

    # Subset the data matrix for these cells
    subset_X = X_dense[indices, :]

    # Calculate the range for the subset
    data_min = np.min(subset_X)
    data_max = np.max(subset_X)

    # Print the range for the current dataset_id
    print(f"Dataset ID {dataset_id}: Range of .X values is {data_min} to {data_max}")
```
